```python
# Import modules
import os
from glob import glob
```

```python
# Navigate to the directory
root  = os.path.join(os.getcwd(), '2. Rearranged Order/')
```

```python
# Sort the file names 
paths = sorted(glob(root+'*.txt'))
```

```python
all_paths = [path.split('\\')[-1] for path in paths]
all_paths
```

```python
# Initialize lengths
lengths = []
```

```python
# Initialize new file:
new_file = 'USC Student Manual and Enrollment Guide.txt'
with open(new_file, 'w'):
  pass
```

```python
# Copy each file to new file
for path in paths:
  with open(new_file, 'a') as out_, open(path, 'r') as in_:
    print(path)
    text = in_.read()
    lengths.append(len(text))
    out_.write(text + '\n\n\n')
```

```python
# Get total text length
sum(lengths)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.lineplot(lengths)
plt.show()
```
