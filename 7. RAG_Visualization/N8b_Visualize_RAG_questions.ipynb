```python
import pandas as pd
from umap.umap_ import UMAP
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load DataFrame from N7
df_euclid = pd.read_csv('embeddings_and_distances(euclid).csv')
df_cosine = pd.read_csv('embeddings_and_distances(cosine).csv')
df_texts_euclid = df_euclid.copy(deep=True)
df_texts_euclid = df_texts_euclid[df_texts_euclid['question'].isna()]
df_texts_cosine = df_cosine.copy(deep=True)
df_texts_cosine = df_texts_cosine[df_texts_cosine['question'].isna()]
```

```python
df_dicts = [{"metric": "Euclidean distance", "df": df_euclid},{"metric": "Cosine Similarity", "df": df_cosine}]
```

```python
# Number of dimensions of embedding per document 
# Note that we use UMAP to reduce this to 2 dimensions
len(df_cosine['embedding'][0][1:-1].split(', '))
```

```python
# Prepare embeddings
# Prepare the embeddings for UMAP
array = [row for row in df_cosine["embedding"]]
array_float = [[float(x) for x in row[1:-1].split(', ')] for row in array]

# Prepare the embeddings for UMAP
embeddings = np.array(array_float)

# Reduce dimensionality with UMAP
reducer = UMAP(random_state = 2)
embedding_reduced = reducer.fit_transform(embeddings)
```

```python
# Using UMAP
# Find the 5 closest vectors
for i in range(1,8):
  for dict in df_dicts:

    # Initialize df and top n closest indices
    df = dict['df']

    # Filter metric
    if dict['metric'] == 'Euclidean distance':
      closest_vectors_indices = df_texts_euclid.nsmallest(5, f'dist_{i}')['id'].values
      question_row = df[df[f"dist_{i}"] == df[f"dist_{i}"].min()]
    else:
      closest_vectors_indices = df_texts_cosine.nlargest(5, f'dist_{i}')['id'].values
      question_row = df[df[f"dist_{i}"] == df[f"dist_{i}"].max()]

    # Display question, most similar articles, and score
    print(f"Top 5 most similar articles for question {i}: {question_row['question'].values[0]} (Metric: {dict['metric']})")
    
    top_n_df = df[df["id"].isin(closest_vectors_indices)].sort_values(by=[f'dist_{i}'],ascending = (dict['metric'] == 'Euclidean distance'))
    for idx, row in top_n_df.iterrows():
      print(f"  Score: {row[f'dist_{i}']:.6f} - {row['source']}")
    
    # Plot the reduced embeddings
    plt.scatter(embedding_reduced[:,  0], embedding_reduced[:,  1], c='gray', alpha=0.2)

    # Highlight the question embedding and the  5 closest vectors
    plt.scatter(embedding_reduced[df["id"].isin(closest_vectors_indices),  0], \
                embedding_reduced[df["id"].isin(closest_vectors_indices),  1], c='red', alpha=1, label='Top 5 articles')
    plt.scatter(embedding_reduced[df["id"] == question_row["id"].values[0],  0], \
                embedding_reduced[df["id"] == question_row["id"].values[0],  1], c='blue', alpha=1, marker='*', label='Question')

    # Add labels and title
    plt.title(f"UMAP Visualization of Text Embeddings with Question {i}")
    plt.xlabel(f"Question: {question_row['question'].values[0]}")
    plt.ylabel(f"Metric: {dict['metric']}")
    plt.legend(loc=2)
    plt.show()
```
